```python
# Experiment 3 Using Cross Validation along with SFS
```

```python
kfolds= StratifiedKFold(n_splits=4,shuffle=True,random_state=42)
train_set,test_set= train_test_split(df,test_size=0.15,stratify=df.status)
train_set= train_set.reset_index(drop=True)
model = LogisticRegression(max_iter=1000)
```

```python
X=train_set[sel_cols]
y=train_set.status
val_scores=[]
for train_idx,valid_idx in kfolds.split(X,y):
    X_tr= X.loc[train_idx]
    y_tr= y.loc[train_idx]
    model.fit(X_tr,y_tr)
    X_val=X.loc[valid_idx]
    y_val=y.loc[valid_idx]
    val_scores.append(model.score(X_val,y_val))
print(val_scores)
```

```python
test_set=test_set.reset_index(drop=True)
X=test_set[sel_cols]
y=test_set.status
model.score(X,y)
X= scaler.fit_transform(X)
y_pred= model.predict(X)
print(confusion_matrix(y,y_pred))
print(roc_auc_score(y,y_pred))
```
